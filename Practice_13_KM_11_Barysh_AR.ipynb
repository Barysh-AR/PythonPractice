У 1939 році друком вийшов роман американського письменика Ернеста Вінсента Райта "Ґетсбі". Особливу увагу публіки він здобув через те, що письменник жодного разу не використав у романі літеру "e" (роман складається з більше ніж 50,000 слів), яка є найбільш вживаною літерою в англійській мові.

Порахуйте кількість входжень кожної літери у текст роману. Виведіть список літер англійського алфавіту, а також інформацію про те, скільки відсотків від усіх літер тексту складає кожна літера алфавіту.

Зауваження:

1. Програма має ігнорувати будь-які символи, окрім літер.

2. Переведіть усі літери у верхній або у нижній регістр.

3. Під час виведення відсортуйте літери у порядку зменшення їхнього відсотка. Літера "e" має стояти на останньому місці у цьому списку (хоча кілька літер e там все ж таки зустрічається).

4. Виведіть перші 5 та останні 5 літер відсортованого списку разом з відсотковими значеннями, що їм відповідають. Відсотки округліть до тисячних.

````python

def e(list_):
    text = ''
    for i in list_:  
        text += i
    text = text.upper()
    return text

def f(text, alphabet='ABCDEFGHIJKLMNOPQRSTUVWXYZ'):
    a = []
    for i in alphabet:
        el = text.count(i)
        a += [el]
    return a

def g(list_):
    a = 0
    for i in list_:
        a += i
    return a

def h(list_, sum_):
    a = []
    for i in list_:
        el = float(i) * 100 / sum_   # round(, 3) 
        a += [el]
    return a

def ij(l, alphabet=list('ABCDEFGHIJKLMNOPQRSTUVWXYZ')):
    l = list(l)
    list_ = []
    r =len(l)
    while (len(list_) != r * 2) :
        a = max(l)
        n =0
        for i in l:
            if i == a:
                list_ += [i]
                list_ += [alphabet[n]]
                alphabet.remove(alphabet[n])
                l.remove(i)
            n +=1
    return list_

def print_statistics(ordered_list, word):
    print('\n' +'{} statistics: \n '.format(word))
    n = 1
    for i in range(5):
        print('letter "{}"'.format(ordered_list[n]), end=' ')
        print('{}% '.format(round(ordered_list[n-1], 3)) + 'of the text')
        n += 2
    print('...')
    n = n + len(ordered_list) -20
    for i in range(5):
        print('letter "{}"'.format(ordered_list[n]), end=' ')
        print('{}% '.format(round(ordered_list[n-1], 3)) + 'of the text')
        n += 2
    return

alphabet='ABCDEFGHIJKLMNOPQRSTUVWXYZ'

file = open('gadsby.txt', 'r')
all_text = file.readlines()
text = e(all_text[110::1])
introductson = e(all_text[:110:1])
all_text = file.read()
all_text = all_text.upper()
file.close()


list_quantities_text = f(text, alphabet)
number_etters_text = g(list_quantities_text)
list_percentages_text = h(list_quantities_text, number_etters_text)

ordered_list_text = ij(list_percentages_text, list(alphabet))
print_statistics(ordered_list_text, 'Novella text')


list_quantities_introductson = f(introductson, alphabet)
number_etters_introductson = g(list_quantities_introductson)
list_percentages_introductson = h(list_quantities_introductson, number_etters_introductson)

ordered_list_introductson = ij(list_percentages_introductson, list(alphabet))
print_statistics(ordered_list_introductson, 'Introductson text')


list_quantities_all_text = []
for i in alphabet:
    el = text.count(i)
    list_quantities_all_text += [el]
number_etters_all_text = g(list_quantities_all_text)
list_percentages_all_text = h(list_quantities_all_text, number_etters_all_text)


ordered_list_all_text = ij(list_percentages_all_text)
print_statistics(ordered_list_all_text, 'All text')
````

Архів archive.zip містить 140 txt-файлів, в яких збережено інфорамцію про те, якими іменами називали дітей у США в період з 1880 по 2019 рік.

Структура рядка кожного файлу така:

*name,sex,number_of_occurrences*

Напишіть програму, яка зчитуватиме кожен файл та визначатиме окремо чоловічі, окремо жіночі імена, що були найпопулярнішими хоча б в один з років. Далі програма повинна підраховувати, скільки разів кожне з цих імен ставало найпопулярнишим, та виводити ранжовані списки чоловічіх та жіночих імен у вигляді:

```
Michael 44
John 44
Robert 17
...

Mary 76
Jennifer 15
Emily 12
...
```

````python

def f(list_of_element, a):
    n = 0
    for i in list_of_element:
        if i == a:
            name = list_of_element[n-1]
            break
        n += 1
    return name

f_list =[]
m_list =[]

year = 1880
while year < 2020:
    file_name = '''archive\yob''' + str(year) + '''.txt'''

    file = open(file_name, 'r')
    text_list = file.readlines()

    list_of_element = []
    for line in text_list:
        elem = ''
        for symbol in str(line):
            # print(symbol)
            if symbol == ',' or symbol == '\n':
                list_of_element += [elem]
                elem = ''
            else:
                elem += str(symbol)

    f_list += [f(list_of_element, 'F')]
    m_list += [f(list_of_element, 'M')]

    year += 1


def g(list_):
    lis = []
    for i in list_:
        if i not in lis:
            lis += [i]
    return lis

list_f = g(f_list)
list_m = g(m_list)

def h(names, list_):
    lis = []
    for i in names:
        lis += [list_.count(i)]
    return lis

def print_result(names, list_):
    a = h(names, list_)
    n = 0
    for i in names:
        print('{} {}'.format(i, a[n]))
        n +=1 
    print('...')

print_result(list_m, m_list)
print_result(list_f, f_list)
````
